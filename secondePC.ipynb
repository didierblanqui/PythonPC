**Import des librairies numpy et matplotlib**

```
import numpy as np
import matplotlib.pyplot as plt
```

**On stocke les positions x et y d'un point dans un tableau numpy**

```
t=np.array([0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5])

x = np.array([0.0,16.2,29.9,38.9,46.6,57.2,71.7,84.8])
y = np.array([15.0,14.95,19.0,26.1,31.3,35.4,35.7,30.0])

n = len(t)
pos=np.zeros((n,2),float)
pos[:,0] = x
pos[:,1] = y

print(pos)
```

**Affichage de la trajectoire**

```
plt.xlim(0,100)
plt.ylim(0,40)
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.axis("equal")
plt.plot(pos[:,0],pos[:,1],"o")
```

**Approcher le vecteur vitesse d'un point à l’aide du vecteur déplacement MM′, où M et M′ sont les positions successives à des instants voisins séparés de Δt**

```
plt.xlim(0,100)
plt.ylim(0,40)
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.axis("equal")
plt.plot(pos[:,0],pos[:,1],"o")
#Tracé des vecteurs déplacement en rouge
for i in range(n-1):
    deplacement = pos[i+1,:] - pos[i,:]
    plt.quiver(pos[i,0],pos[i,1],deplacement[0],deplacement[1],angles='xy', scale_units='xy',scale=1,color='red')
```

**On rajoute l'opposé du vecteur déplacement précédent en jaune**

```
plt.xlim(0,100)
plt.ylim(0,40)
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.axis("equal")
plt.plot(pos[:,0],pos[:,1],"o")
#Tracé des vecteurs déplacement en rouge
for i in range(n-1):
    deplacement = pos[i+1,:] - pos[i,:]
    plt.quiver(pos[i,0],pos[i,1],deplacement[0],deplacement[1],angles='xy', scale_units='xy',scale=1,color='red')
#Tracé de l'opposé des vecteurs déplacement en jaune (à la suite du rouge)
for i in range(1,n-1):
    deplacement2 = pos[i-1,:] - pos[i,:]
    plt.quiver(pos[i+1,0],pos[i+1,1],deplacement2[0],deplacement2[1],angles='xy', scale_units='xy',scale=1,color='yellow')
```

**On trace la différence des deux vecteurs**

```
plt.xlim(0,100)
plt.ylim(0,40)
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.axis("equal")
plt.plot(pos[:,0],pos[:,1],"o")
#Tracé des vecteurs déplacement en rouge
for i in range(n-1):
    deplacement = pos[i+1,:] - pos[i,:]
    plt.quiver(pos[i,0],pos[i,1],deplacement[0],deplacement[1],angles='xy', scale_units='xy',scale=1,color='red')
#Tracé de l'opposé des vecteurs déplacement en jaune (à la suite du rouge)
for i in range(1,n-1):
    deplacement2 = pos[i-1,:] - pos[i,:]
    plt.quiver(pos[i+1,0],pos[i+1,1],deplacement2[0],deplacement2[1],angles='xy', scale_units='xy',scale=1,color='yellow')
#Tracé des variations de vecteurs déplacement
for i in range (1,n-1):
    deltav = (pos[i+1,:] - pos[i,:])-(pos[i,:] - pos[i-1,:])
    plt.quiver(pos[i,0],pos[i,1],deltav[0],deltav[1],angles='xy', scale_units='xy',scale=1,color='blue')

plt.show() # affiche la figure a l'ecran
```

**Finalement on ne garde que les vecteurs variations de vitesse moyenne**

```
plt.xlim(0,100)
plt.ylim(0,40)
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.axis("equal")
plt.plot(pos[:,0],pos[:,1],"o--")
#Tracé des variations de vecteurs déplacement
for i in range (1,n-1):
    deltav = (pos[i+1,:] - pos[i,:])-(pos[i,:] - pos[i-1,:])
    plt.quiver(pos[i,0],pos[i,1],deltav[0],deltav[1],angles='xy', scale_units='xy',scale=1,color='blue')

plt.show() # affiche la figure a l'ecran
```
